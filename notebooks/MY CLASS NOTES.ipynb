### 2019 Q1

Q1. Compute the semantic representation of the sentence: "Every knight seeks a dragon", whose abstract syntax is given by the following term: `seek (a dragon) (every knight)`

```python
%%lamb
||find|| = find_<e,<e,t>>
||Try|| = try_<e,<<e,t>,t>>
# ^ these first two will not actually be used

||every|| = λ m_<e,t>: λ n_<e,t>: Forall x_e: (m(x) >> n(x))
||a|| = λ m_<e,t>: λ n_<e,t>: Exists x_e: (m(x) & n(x))

# simple definition of 'seek'
||seek_simple|| = λ o_<<e,t>,t>: λ s_<<e,t>,t>: s(λ x_e: o(λ y_e: seek_<e,<e,t>>(x)(y)))

# complex definition of 'seek', as given in 2019 exam
||seek_TF|| = λ o_<<e,t>,t>: λ s_<<e,t>,t>: s(λ x_e: o(  λ y_e: try_<e,<<e,t>,t>> (x) (λ z_e: find_<e,<e,t>>(z)(y))  ))

||knight|| = knight_<e,t>
||dragon|| = dragon_<e,t>
```

```python
AD = a*dragon
EK = every*knight

S_simple_AD_EK = (seek_simple*AD)*EK
S_tf_AD_EK = (seek_TF*AD)*EK
```

### 2019 Q2

Q2. Which reading do you obtain? Give a context that favors this reading.

We obtain the de dicto (de dicto: "what is said") reading. 
Note that Forall quantifier has higher scope than Exists quantifier. Thus we are not talking about one specific something. 

*Note: To be talking about one specific something (de re: "a thing"), we would need Existential quantifier to be higher scope.*

As for context to favor this de dicto reading, in my opinion this is the more obvious reading. 
A valid context might be something like... there are a lot of knights and they are all looking for their own dragon so they can get famous for finding (killing?) a new one.

### 2019 Q3

Q3:
Give two alternative semantic recipes for seek that lead to two other possible readings. Explain these two additional readings. Give contexts that favor them.

```python
'''First, lets try just switching up the order of arguments for seek, 
without changing the definition from before.'''
sent = (seek_TF*(EK))*AD
sent.tree()
```

In the output from this, we see that the Exists quantifier is higher in scope now. 
So we now have a single dragon. 
However, this sentence means "There is a dragon that seeks every knight." 
(Since we switch the order of arguments without changing definition of 'seeks'.)

Based on this, it seems that by simply swapping the order that 'seek' expects its obj/subj args,
we could have Exists quantifier higher in scope.

Note that we want to have a reading with both:
(1) Exists quantifier high in scope, and 
(2) correct obj/subj for "seeks(knight,dragon)"

Assuming we want to keep the application order `seek (a dragon) (every knight)` (since that's what we were told is the proper order, in the exam instructions), we will not change the order of o/s in the lambda bindings.
Instead, we will change the order of o/s inside the interior of the definition.

```python
%%lamb
# definition of 'seek' with s and o swapped
||seek_TF_so1|| = λ o_<<e,t>,t>: λ s_<<e,t>,t>: o(λ x_e: s(  λ y_e: try_<e,<<e,t>,t>> (x) (λ z_e: find_<e,<e,t>>(z)(y))  ))
```

```python
# original complex definition of 'seek', as given in 2019 exam.
# Exactly same as before, just repeated here for reference.
seek_TF
```

```python
'''The output of this is almost right, but, it means "There's a dragon that seeks every knight",
which is not what we want.'''
sent = (seek_TF_so1*(AD))*EK
sent.tree()
```

We need the x in dragon(x) and the x in the first arg of try(x)(...) to NOT be the same. 
Let's try also swapping the lambda bindings for x and y in the definition. 

```python
%%lamb
# definition of 'seek' with s and o swapped AND the lambda y and x swapped
||seek_TF_so2|| = λ o_<<e,t>,t>: λ s_<<e,t>,t>: o(λ y_e: s(  λ x_e: try_<e,<<e,t>,t>> (x) (λ z_e: find_<e,<e,t>>(z)(y))  ))
```

```python
seek_TF # the original again for reference
```

```python
'''The output of this is what we want. '''
sent = (seek_TF_so2*(AD))*EK
sent.tree()
```
